```python
def get_wordcount(review):
    selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
    selected_word = {'awesome':0, 'great':0, 'fantastic':0, 'amazing':0, 'love':0, 'horrible':0, 'bad':0, 'terrible':0, 'awful':0, 'wow':0, 'hate':0}
    rev=review.split()
    for w in selected_words:
        count=0
        count=rev.count(w)
        selected_word[w]=count
    return selected_word
```

```python
productssm=graphlab.SFrame('amazon_baby.gl/')
productssm['word_count']=productssm['review'].apply(get_wordcount)
productssm=productssm[productssm['rating']!=3]
productssm['sentiment']=productssm['rating']>=4
train_data_select_model,test_data_select_model=productssm.random_split(.8,seed=0)
selected_words_model = graphlab.logistic_classifier.create(train_data_select_model,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data_select_model)
```

```python
selected_words_model.evaluate(test_data_select_model)
```
